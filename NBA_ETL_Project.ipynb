```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "Resources/NBA_player_of_the_week.csv"
weekly_player_df = pd.read_csv(csv_file)
weekly_player_df.head()
```

```python
csv_file2 = "Resources/NBA_season1718_salary.csv"
salary_df2 = pd.read_csv(csv_file2)
salary_df2.head()
```

```python
#drop rows
clean_weekly_player_df = pd.DataFrame(weekly_player_df.loc[weekly_player_df["Season"]=="2017-2018"])
clean_weekly_player_df
```

```python
#make a copy
cleaner_weekly_player_df = cleaner_weekly_player_df.copy()
cleaner_weekly_player_df
```

```python
#drop columns
cleaner_weekly_player_df = clean_weekly_player_df[['Player','Team','Position','Season']]
cleaner_weekly_player_df
```

```python
#make a copy
clean_salary_df = salary_df2.copy()
clean_salary_df
```

```python
#dropped 'Unnamed:0' column
new_salary_df = clean_salary_df[['Player','Tm','season17_18']]
new_salary_df
```

```python
#rename columns
renamed_salary_df = new_salary_df.rename(columns={"Tm": "Team Abv.",
                                                         "season17_18": "Salary"})
renamed_salary_df
```

```python
#renamed columns
renamed_weekly_player_df = cleaner_weekly_player_df.rename(columns={"Team": "Team Name"})
renamed_weekly_player_df
```

```python
#re-indexed POTW table
indexed_weekly_player_df = renamed_weekly_player_df.reset_index()
indexed_weekly_player_df
```

```python
#dropped old index column
indexed_player_df = indexed_weekly_player_df[['Player','Team Name','Position','Season']]
indexed_player_df
```

```python
#clarifying table data
renamed_salary_df
```

```python
#format salary number - ended up not using this because it turned the INT into a STRING
#clean_salary_df = renamed_salary_df.style.format({"Salary": "{:,.0f}"})
#clean_salary_df
```

```python
#exported table
renamed_salary_df.to_csv('NBA_salary.csv')
```

```python
#exported to CSV
indexed_player_df.to_csv('NBA_players_of_week.csv')
```

```python
#connecting to SQL
connection_string = "postgres:XxXxX@localhost:5432/NBA2_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
#change column titles to lowercase
matching_salary_df = renamed_salary_df.rename(columns={"Player": "player",
                                 "Team Abv.": "team_abv", "Salary": "salary"})
matching_salary_df
```

```python
#change column titles to lowercase
indexed_player2_df = indexed_player_df.rename(columns={'Player':'player', 'Team Name':'team_name', 
                                                        'Position':'position', 'Season':'season'})
indexed_player2_df
```

```python
#check table names
engine.table_names()
```

```python
#FINAL TABLE EXPORT TO SQL
matching_salary_df.to_sql(name='nba_salary3', con=engine, if_exists='append', index=True)
```

```python
#FINAL TABLE EXPORT TO SQL
indexed_player2_df.to_sql(name='nba_players_of_week3', con=engine, if_exists='append', index=True)
```
